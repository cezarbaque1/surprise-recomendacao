```python
from conn.apis import *
import pandas as pd

retorno = get_all_respostas()
getDf = pd.DataFrame(retorno)
getDf.head()
```

```python
print(len(getDf))
```

```python
df = getDf.explode('produtos')
df.head()
```

```python
df['produtosSeparado'] = df['produtos'].apply(lambda x:x['produto'])
df['respostaProduto'] = df['produtos'].apply(lambda x:x['response'])
df.head()
```

```python
grouped_df = df.groupby('respostaProduto').agg(
    count=('respostaProduto', 'count')
).sort_values(by='count', ascending=False).reset_index()
grouped_df
```

```python
dfAjustado = df[df['respostaProduto'] == 'Amei!' ]
print(len(dfAjustado))
dfAjustado.head()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python
#Separa variaveis sucessoras e predecessoras
x = dfAjustado[['1','2','3','4','5']]
y = dfAjustado['produtosSeparado']

#transforma o y em um array
y = y.to_numpy()

# Dividir os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)

print(len(x_treino))
print(len(x_teste))
x_treino.head()
```

```python
#Criar a Pipeline
encoder = OneHotEncoder()
modelo = GaussianNB()

pipeline = Pipeline([
    ('encoder', ColumnTransformer([('encoder', encoder, ['1','2','3','4','5']) ] ) ),
    ('modelo', modelo)
])
```

```python
pipeline.fit(x_treino, y_treino)
```

```python
predictions = pipeline.predict(x_teste)
print(predictions)
```

```python
accuracy = pipeline.score(x_teste, y_teste)
print(f"Acurácia da pipeline: {accuracy}")
```

```python
x_teste['idProduto'] = predictions
x_teste.head()
```
