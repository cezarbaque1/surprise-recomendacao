```python
from conn.apis import *
import pandas as pd

retorno = get_all_respostas()
getDf = pd.DataFrame(retorno)
getDf.head()
```

```python
print('Número de Respostas: ' + str(len(getDf)))
```

```python
df = getDf.explode('produtos')
print('Número de Produtos Avaliados: ' + str(len(df)))
df.head()
```

```python
df['produtosSeparado'] = df['produtos'].apply(lambda x:x['produto'])
df['respostaProduto'] = df['produtos'].apply(lambda x:x['response'])
df = df[df['respostaProduto'] != 'Não Gostei' ]
df.head()
```

```python
gostei_df = df.groupby('respostaProduto').agg(
    count=('respostaProduto', 'count')
).sort_values(by='count', ascending=False).reset_index()
gostei_df
```

```python
grouped_df = df.groupby('produtosSeparado').agg(
    count=('produtosSeparado', 'count')
).sort_values(by='count', ascending=False).reset_index()
grouped_df.head()
```

```python
countdf = grouped_df
countdf['contagem'] = countdf['count']
countdf = countdf[['contagem', 'produtosSeparado']]
countdf = grouped_df.groupby('contagem').agg(count=('produtosSeparado', 'count')
).sort_values(by='count', ascending=False).reset_index()
countdf
```

```python
dfAjustado = df[df['respostaProduto'] != 'Não Gostei' ] #Filtras tudo que não é "Não Gostei"

#Filtras produtos que tiveram duas respostas ou mais
dfAjustado = dfAjustado.merge(grouped_df,how='left', on='produtosSeparado') 
dfAjustado = dfAjustado[dfAjustado['count'] > 1]

#Duplicar linhas que foram selecionadas como "Amei"
dfAmei = dfAjustado[dfAjustado['respostaProduto'] != 'Amei!' ] 
dfAjustado = pd.concat([dfAjustado, dfAmei], ignore_index=True).reset_index()

#Resultado
print(len(dfAjustado))
dfAjustado.head()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python
#Separa variaveis sucessoras e predecessoras
x = dfAjustado[['1','2','3','4','5']]
y = dfAjustado['produtosSeparado']

#transforma o y em um array
y = y.to_numpy()

# Dividir os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)

print(len(x_treino))
print(len(x_teste))
x_treino.head()
```

```python
#Criar a Pipeline
encoder = OneHotEncoder()
modelo = GaussianNB()

pipeline = Pipeline([
    ('encoder', ColumnTransformer([('encoder', encoder, ['1','2','3','4','5']) ] ) ),
    ('modelo', modelo)
])
```

```python
pipeline.fit(x_treino, y_treino)
```

```python
predictions = pipeline.predict(x_teste)
print(predictions)
```

```python
accuracy = pipeline.score(x_teste, y_teste)
print(f"Acurácia da pipeline: {accuracy}")
```

```python
x_teste['idProduto'] = predictions
x_teste['idProdutoOriginal'] = y_teste
x_teste_prod = x_teste
x_teste_prod.head()
```
